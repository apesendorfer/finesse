```python
# Compare two words

vec1 = model.get_word_vector("test")
vec2 = model.get_word_vector("exam")

magnitude1 = 0;
for i in range(0, 100):
    magnitude1 += (vec1[i] ** 2)
magnitude1 = magnitude1 ** 0.5
print(magnitude1)

magnitude2 = 0;
for i in range(0, 100):
    magnitude2 += (vec2[i] ** 2)
magnitude2 = magnitude2 ** 0.5
print(magnitude2)

def dot(K, L):
   if len(K) != len(L):
      return 0

   return sum(i[0] * i[1] for i in zip(K, L))

print(dot(vec1, vec2) / magnitude1 / magnitude2)
```
