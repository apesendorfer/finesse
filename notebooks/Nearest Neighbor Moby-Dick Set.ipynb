```python
import fasttext
import numpy as np
```

```python
# cosine similarity of word vectors

from numpy import dot
from numpy.linalg import norm

# cos_sim = dot(a, b)/(norm(a)*norm(b))
```

```python
# Set of words from Herman Melville's Moby-Dick TF-IDF Analysis

f = open("moby-dick_tfidf_results.txt", "r")

doc = f.read()

f.close()

#print(doc)

words = doc.split(", ")

print(words)

# lines = f.readlines()

# f.close()

# words = []

# for line in lines:
#     if line.isupper():
#         #print(line)
#         new_word = line.lower().strip()
#         if new_word not in words:
#             words.append(new_word)
        
# print(words[0:100])
```

```python
print(len(words))
```

```python
#create numpy array to store word vectors

#currently set to store top 1000 words as their 100 dimensional word vectors

arr = np.ndarray(shape=(12296, 300))
```

```python
#load fasttext model

model = fasttext.load_model("/Users/alex/result/cc.en.300.bin")
```

```python
# Store word vectors for the 12296 words in numpy array

for i in range(0, 12296):
    arr[i] = model.get_word_vector(words[i])
```

```python
# user word
user_word = "pleasant"

user_word_vec = model.get_word_vector(user_word)

# iterate through numpy array and find vector with greatest cosine similarity
index = -1
cos_sim = -1
for i in range(0, 12296):
    current = dot(user_word_vec, arr[i]) / (norm(user_word_vec) * norm(arr[i]))
    if current > cos_sim and words[i] != user_word and user_word[0:2] != words[i][0:2] and user_word[0:3] not in words[i]:
        cos_sim = current
        index = i

print(words[index] + " " + str(cos_sim))

print(model.get_nearest_neighbors(user_word)[0])
```

## Possibly promising results

```python
# user word
user_word = "dark"

user_word_vec = model.get_word_vector(user_word)

results = {}

# iterate through numpy array and find vectors that meet a threshold for cosine similarity
cos_sim = 0.55
for i in range(0, 12296):
    current = dot(user_word_vec, arr[i]) / (norm(user_word_vec) * norm(arr[i]))
    if current > cos_sim and words[i] != user_word and user_word[0:2] != words[i][0:2] and user_word[0:3] not in words[i]:
        results[words[i]] = current
```

```python
print(model.get_nearest_neighbors(user_word)[0:5])
```

```python
print(results)
```

# Below, I conducted a brief runtime analysis

```python
import timeit

start = timeit.default_timer()

# user word
user_word = "mean-spirited"

user_word_vec = model.get_word_vector(user_word)

results = {}

# iterate through numpy array and find vectors that meet a threshold for cosine similarity
cos_sim = 0.6
for i in range(0, 12296):
    current = dot(user_word_vec, arr[i]) / (norm(user_word_vec) * norm(arr[i]))
    if current > cos_sim and words[i] != user_word and user_word[0:2] != words[i][0:2] and user_word[0:3] not in words[i]:
        results[words[i]] = current

stop = timeit.default_timer()

print('Time: ', stop - start)
```
