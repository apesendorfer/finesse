```python
# Texts included in word list

# gatsby, moby-dick, kerouac, gbg, invman, frankenstein

txt1 = "gatsby.txt"
txt2 = "moby-dick.txt"
txt3 = "kerouac.txt"
txt4 = "gbg.txt"
txt5 = "invman.txt"
txt6 = "frankenstein.txt"

lst1 = [txt1, txt2, txt3, txt4, txt5, txt6]

def concatenate(lst):
    s = ""
    for text in lst:
        f = open(text, "r")
        temp = f.read()
        f.close()
        s = s + temp
    return s

s = concatenate(lst1)
```
