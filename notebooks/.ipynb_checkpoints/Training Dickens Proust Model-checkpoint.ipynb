```python
import fasttext
```

```python
f = open("data.txt", "r")

text = f.read()

f.close()
```

```python
print(text)
```

```python
documentA = text.split()

for i in range(0, len(documentA)):
    documentA[i] = documentA[i].strip(".,!?;:\'\"\[]*#()$_-").lower()

print(documentA[0:100])

text = " ".join(documentA)

print(text[0:1000])
```

```python
f = open("dickensproustoutput.txt", "w")

f.write(text)

f.close()
```

```python
model = fasttext.train_unsupervised('dickensproustoutput.txt', "cbow", thread=4)
```

```python
model3 = fasttext.train_unsupervised('dickensproustoutput.txt', "skipgram", thread=4)
```
