```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# conda install -c conda-forge scikit-learn
```

```python
f = open("kerouac.txt", "r")

documentA = f.read().split()

f.close()

for i in range(0, len(documentA)):
    documentA[i] = documentA[i].strip(".,!?;:\'\"")

print(documentA[0:100])

documentA = " ".join(documentA)

print(documentA[0:1000])
```

```python
f = open("gutenberg_dict.txt", "r")

documentB = f.read().split()

f.close()

for i in range(0, len(documentB)):
    documentB[i] = documentB[i].strip(".,!?;:\'\"\(\)")

print(documentB[0:100])

documentB = " ".join(documentB)

print(documentB[0:1000])
```

```python
print(documentB[0:1000])
```

```python
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform([documentA, documentB])
feature_names = vectorizer.get_feature_names()
dense = vectors.todense()
denselist = dense.tolist()
df = pd.DataFrame(denselist, columns=feature_names)
```

```python
maxtfidf = -1

result = ""

for word in df:
    if df.at[0, word] > maxtfidf:
        result = word
        maxtfidf = df.at[0, word]
```

```python
print(result)
print(maxtfidf)
```

```python
from nltk.corpus import stopwords
# stopwords.words('english')
```

```python
threshold = 0.005
0.001

result = []

for word in df:
    if df.at[0, word] > threshold and word not in stopwords.words('english'):
        result.append(word)
```

```python
print(result)
```

```python
lower = 0.003
upper = 0.0002


result = []

for word in df:
    if df.at[0, word] > lower and df.at[1, word] < upper and word not in stopwords.words('english'):
        result.append(word)
```

```python
print(result)
```

```python
lower = 0.003
upper = 0.0002


result = []

for word in df:
    if df.at[0, word] > lower and df.at[1, word] < upper and word not in stopwords.words('english'):
        result.append(word)
```

```python
result = []

for word in df:
    if df.at[0, word] > df.at[1, word] and word not in stopwords.words('english') and not word.isnumeric():
        result.append(word)
```

```python
print(result)
```

```python
print(len(result))
```

```python
f = open("kerouac_tfidf_results.txt", "w")

f.write(", ".join(result))

f.close()
```
