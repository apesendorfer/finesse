```python
# https://towardsdatascience.com/natural-language-processing-feature-engineering-using-tf-idf-e8b9d00e7e76

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from nltk.corpus import stopwords
# stopwords.words('english')
```

```python
# Set of words from Herman Melville's Moby-Dick TF-IDF Analysis

f = open("hitchhikers2.txt", "r")

documentA = f.read()

f.close()

#print(documentA)

documentA = documentA.split()

#print(documentA)

for i in range(0, len(documentA)):
    documentA[i] = documentA[i].strip(".,!?;:\'\"\[]*#()$").lower()

print(documentA[0:100])

documentA = " ".join(documentA)

print(documentA[0:1000])
```

```python
f = open("gutenberg_dict.txt", "r")

documentB = f.read().split()

f.close()

for i in range(0, len(documentB)):
    documentB[i] = documentB[i].strip(".,!?;:\'\"\(\)")

print(documentB[0:100])

documentB = " ".join(documentB)

print(documentB[0:1000])
```

```python
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform([documentA, documentB])
feature_names = vectorizer.get_feature_names()
dense = vectors.todense()
denselist = dense.tolist()
df = pd.DataFrame(denselist, columns=feature_names)
```

```python
words = []

for word in df:
    if df.at[0, word] > 10 * df.at[1, word] and word not in stopwords.words('english') and not word.isnumeric():
        words.append(word)
```

```python
words = []

for word in df:
    if df.at[0, word] > df.at[1, word] and word not in stopwords.words('english') and not word.isnumeric():
        words.append(word)
```

```python
#create numpy array to store word vectors

#currently set to store top 10405 words as their 100 dimensional word vectors

arr = np.ndarray(shape=(10405, 300))
```

```python
#load fasttext model

model = fasttext.load_model("/Users/alex/result/cc.en.300.bin")
```

```python
# Store word vectors for the 10405 words in numpy array

for i in range(0, 10405):
    arr[i] = model.get_word_vector(words[i])
```

```python
# user word
user_word = "pleasant"

user_word_vec = model.get_word_vector(user_word)

# iterate through numpy array and find vector with greatest cosine similarity
index = -1
cos_sim = -1
for i in range(0, 10405):
    current = dot(user_word_vec, arr[i]) / (norm(user_word_vec) * norm(arr[i]))
    if current > cos_sim and words[i] != user_word and user_word[0:2] != words[i][0:2] and user_word[0:3] not in words[i]:
        cos_sim = current
        index = i

print(words[index] + " " + str(cos_sim))

print(model.get_nearest_neighbors(user_word)[0])
```
