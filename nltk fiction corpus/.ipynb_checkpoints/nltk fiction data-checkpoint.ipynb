```python
f = open("data.txt", "w")

for text in nltk.corpus.gutenberg.fileids():
    for word in nltk.Text(nltk.corpus.gutenberg.words(text)):
        f.write(word.strip("!@#$%^&*\(\)\[\].,-<>?/\"\';:") + " ")
        
f.close()
```
